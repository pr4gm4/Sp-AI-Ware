#### Importing Modules

```python
import pandas as pd
import warnings
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings("ignore")
```

#### Loading Dataset

```python
size = 50
dataset = pd.read_csv("../../Datasets/"+str(size)+"dataset.csv")
dataset
```

#### Normalization

```python
#normalizza il dataset utilizzando la tecnica MinMax
mmscaler = MinMaxScaler()
dataset.iloc[:,2:] = mmscaler.fit_transform(dataset.iloc[:,2:])
dataset
```

#### Standardization

```python
#Standardizza il dataset togliendo le features non numeriche
scaler = StandardScaler()
dataset.iloc[:,2:] = scaler.fit_transform(dataset.iloc[:,2:])
dataset
```

#### Feature Filtering

```python
#crea una matrice di correlazione di spearman per capire quali sono le features piu' correlate a is_malicious
correlation = dataset.corr("spearman")
correlation
```

```python
#ordina le features della prima riga in ordine discendente di correlazione da is_malicious
best_features = correlation.iloc[0]
best_features = best_features.abs()
best_features = best_features.sort_values(ascending=False)
best_features = best_features[0:26]
best_index = best_features.index
best_index
```

```python
#reinserisce il valore sha corrispondente per ogni riga
ref_dataset = dataset.loc[:,best_index]
ref_dataset["sha"] = dataset.loc[:,"sha"]
ref_dataset
```

#### Export Filtered Dataset

```python
file_path = "../../Datasets/"+str(size)+"dataset.csv"
ref_dataset.to_csv(file_path,index=False)
```
